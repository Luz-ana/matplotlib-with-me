```python
import matplotlib.pyplot as plt
import numpy as np 
```

# lets know about the what the term scatter  it just some points that are scattered on x axis and y axis  and shows relationship between two terms lets understand with an example

```python
study = np.array([1,2,3,4,5])
marks = np.array([10,20,30,40,100])
plt.scatter(study,marks)
plt.show()
```

* as you can see on the graph mentioned above more the study hours more you get marks on subjects  so in this way we can simply analyze the relationship between two factors
* but we might ask bar also shows such relation but it helps to compare between two person who got more marks whereas scatter helps to analyze the relationship between two factor
* from above graph we can also that we have positive correlation between the variables because when one variable increases next is also increasing => positive correlation
#Lets try to fit a straight line that best fits the given condition we already have the actual values of y but we find the predicted value that best fits the stright line when given value of x using y = mx+b and m is the slope and b is the y intercept

```python
study = np.array([1,2,3,4,5])
marks = np.array([10,20,30,40,100])
plt.scatter(study,marks)
m,b = np.polyfit(study,marks,1)
plt.plot(study,m*study+b,"red")
plt.show()
```

above line shows how y value changes when x change in visual format so we dont have to see each point one by one and its a reference actual value of y can change it can be abit higher or lower
