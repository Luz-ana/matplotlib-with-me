#HISTOGRAM => this simply helps us to find number of peoples under particular range of data 


```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
no = np.random.randint(10,20,(15)) #last values is for no of data we need . If we want the size thing we need to have arange
print(no)
```

```python
data = [19,17,13,14,11,19,16,16,11,17,16,18,11,19,14]
plt.hist(data)
```

* so the histogram displayed above is not showing exact frequency based figure now lets set the above fig so it can show us frequency based  using bins
* 

```python
bins = [10,13,16,19]
plt.xlabel("data",fontsize=12)
plt.hist(data,bins,color = "pink",edgecolor = "black")
```

```python
data = [19,17,13,14,11,19,16,16,11,17,16,18,11,19,14]
plt.hist(data,cumulative = "true") #cumulative adds up the frequency with the previous one suppose we have freq 1 for 19 then freq for 17 is 1+2=3 like this 
```

```python
data = [19,17,13,14,11,19,16,16,11,17,16,18,11,19,14]
plt.hist(data,bottom=[10]) #this sets the starting point of the vertical(yaxis)
```

```python
#lets set the style of our histogram 
data = [19,17,13,14,11,19,16,16,11,17,16,18,11,19,14]
plt.hist(data,histtype = "step")
#
```

```python
data = [19,17,13,14,11,19,16,16,11,17,16,18,11,19,14]
plt.hist(data,histtype = "step")
plt.hist(data,histtype = "stepfilled",color = "pink",edgecolor = "black") #in this way we can fill the boxes and show in steps 
```

* if you want to highlight a particular x value we can use a line to represent it 

```python
#axislinevertical
axvline = 16.5
#plt.hist(data,axvline,color = "b") only accepts bins data color not histogram parameter
plt.hist(data,label = "pinky",color = "pink") # only mentioning label here wont give us any value we need legend
plt.axvline(16.5,color = "black",label ="line")
plt.legend()
plt.show()
```
