```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.plot(x,y,color = "pink")
plt.show()
```

```python
# bar graph of my study routine 
df = pd.DataFrame({"subjects":[".net","ecommerce","software engineering","compiler","technical writing","egovernance"],"time":[1,2,3,6,1,2]})
plt.bar(df["subjects"],df["time"],width = 0.2)
plt.xlabel("subject")
plt.ylabel("time")
plt.xticks(rotation = 45) #this creates rotation on the terms below the bar chart 
plt.show()
```

```python
#  to show which subject has taken max time we can simply use 
ttime = np.max(df["time"])
subject =  df[df["time"] == ttime]["subjects"]
print(f"{subject.values[0]}  ")
```

```python
#colors = ["red" if t== ttime for t in df["time"]] thats a mistake because we must use else in the comprehensive list 
colors = ["red" if t== ttime else "yellow" for t in df["time"]]
plt.bar(df["subjects"],df["time"],color = colors)
plt.xticks(rotation=45)
plt.xlabel("subject",fontsize = 10)
plt.ylabel("time", fontsize = 11)
plt.title("My study routine")
plt.show()
```

```python
# so if i want tp compare my routine and my friends routine in visual form lets see how to do that 
b_width = 0.2
df = pd.DataFrame({"subjects":[".net","ecommerce","software engineering","compiler","technical writing","egovernance"],"time":[1,2,3,6,1,2]})
ttime = np.max(df["time"])
colors = ["red" if t== ttime else "yellow" for t in df["time"]]
df1 = pd.DataFrame({"subjects":[".net","ecommerce","software engineering","compiler","technical writing","egovernance"],"time":[10,12,23,16,11,12]})
ttime1 = np.max(df1["time"])
colors1 = ["red" if t== ttime1 else "yellow" for t in df1["time"]]

#plt.bar(df1["subjects"]-b_width/2,df1["time"],width = b_width)
#plt.bar(df["subjects"]+b_width/2,df["time"],width = b_width) these lines give error because inorder to show sideways we need to have a numeric vallue but the df["subjects"] give strings
x = np.arange(len(df["subjects"]))

plt.bar(x-b_width/2,df["time"],width = b_width,color = colors) #its the left side bar graph
plt.bar(x+b_width/2,df1["time"],width = b_width ,color = colors1) #its the right side bar graph 
plt.xlabel("subject")
plt.ylabel("time")
plt.xticks(rotation = 45) #this creates rotation on the terms below the bar chart 
plt.xlabel("subject",fontsize = 10)
plt.ylabel("time", fontsize = 11)
plt.title("Our study routine")
plt.show()
```
